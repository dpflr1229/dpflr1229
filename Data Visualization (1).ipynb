```python
# 데이터 불러오기
insurance = pd.read_csv('insurance.csv')
```

```python
# 데이터 확인
insurance.head(5)
```

```python
# scatterplot 그리기
plt.rcParams['figure.figsize']=(10,5)
sns.scatterplot(data=insurance,x='bmi',y='charges',hue = 'smoker')
```

```python
# regplot 그리기 (추세선)
sns.regplot(data=insurance,x='bmi',y='charges')
```

```python
# lmplot 그리기 (regplot은 hue 파라미터가 없음)
sns.lmplot(data=insurance,x='bmi',y='charges',hue='smoker')
```

regplot과 비교해서 lmplot은 hue를 지정한 칼럼의 추세를 확인할 수 있다 

```python
# swarmplot 그리기
plt.rcParams['figure.figsize']=(7,5)
sns.swarmplot(data=insurance,x='smoker',y='charges')
```

```python
#데이터 불러오기
iris = pd.read_csv('iris.csv',index_col='Id')
```

```python
# 데이터 확인하기
iris.head(5)
```

```python
# histogram 그리기
sns.distplot(iris['Petal Length (cm)'],kde=False)
```

```python
# kdeplot 그리기
sns.kdeplot(iris['Petal Length (cm)'],shade=True)
```

```python
# 2D kdeplot 그리기 
sns.jointplot(data=iris,x='Petal Length (cm)',y='Sepal Width (cm)',kind='kde')
```

jointplot은 수치형 변수간의 상관관계를 그려주는 plot으로 종류에는 scatter,reg,kde,resid,hex 등이 있다

```python
# species별로 분리된 iris 데이터를 가져오기
iris_set = pd.read_csv('iris_setosa.csv', index_col="Id")
iris_ver = pd.read_csv('iris_versicolor.csv', index_col="Id")
iris_vir = pd.read_csv('iris_virginica.csv', index_col="Id")
```

```python
# 분리된 iris데이터를 distplot으로 한눈에 확인
sns.distplot(a=iris_set['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=iris_ver['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_vir['Petal Length (cm)'], label="Iris-virginica", kde=False)
plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
# 분리된 iris데이터를 kdeplot으로 한눈에 확인
sns.kdeplot(data=iris_set['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir['Petal Length (cm)'], label="Iris-virginica", shade=True)
plt.title("Distribution of Petal Lengths, by Species")
plt.legend()
```
