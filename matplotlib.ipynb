```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 기본 그래프 그리기

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

plot과 show 함수를 이용해 시각화를 할 수 있고 위 예시처럼 x,y값을 정해줄 수 있다.

```python
# 스타일 지정
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')
plt.axis([0,6,0,20])
plt.show()
```

plot 함수 가장 마지막에 **포맷 문자열** 을 입력해 그래프 형태를 정할 수 있고(마커)

axis 함수로 x축과 y축의 범위를 설정할 수 있다 [xmin, xmax, ymin, ymax]

axis 함수는 ('on' | 'off' | 'equal' | 'scaled' | 'tight' | 'auto' | 'normal' | 'image' | 'square') 등 다양한 옵션 제공

```python
# 여러개의 그래프 그리기

# 200ms 간격으로 균일하게 샘플된 시간
t = np.arange(0., 5., 0.2)

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, '--', t, t**2, 's', t, t**3, '^')
plt.show()
```

x축 y축 포맷문자열 순으로 나열하면 여러개의 그래프를 한번에 그릴 수 있다.

```python
# 축 레이블 설정 

data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

plot 함수에 data= ~~ 을 추가하여 레이블이 있는 데이터를 그래프로 나타낼 수 있다

또한 plt.xlabel과 plt.ylabel로 축 이름을 설정할 수 있다

이외에도 ‘family’(글씨체), ‘color’, ‘weight’(두께), ‘size’를 각각 지정해주면 폰트를 바꿀 수 있다.

```python
# 범례 표시하기

data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot('data_x','data_y',data=data_dict,label='Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

plot함수에 범례로 사용할 문자를 label로 추가한 뒤 legend 함수를 호출하면 범례가 나타난다

추가로 legend 함수에 loc을 이용하여 위치를 바꿀 수 있다. 

```python
# 선 종류 지정
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot('data_x','data_y','--',data=data_dict,label='Price')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

간단하게 포맷 문자열처럼 ‘ - ‘ (Solid), ‘ - - ‘ (Dashed), ‘ : ‘ (Dotted), ‘ -. ‘ (Dash-dot) 을 이용해 선 종류를 지정할 수 있다

```python
# 색상 지정하기
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}
plt.plot('data_x','data_y','r',data=data_dict)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

포맷문자열로 색상을 입력하면 선의 색상을 지정할 수 있다. (r,g,b,c,m,y,k,w)

다양한 색상은 color='' 을 직접 입력하여 지정할 수 있다

색상과 마커와 선종류를 이어붙이면 동시에 표현할 수 있다 ex) 'ro--'

```python
# 축 스케일 지정
x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x,y)
plt.xscale('linear')
plt.yscale('symlog')
plt.show()
```

xscale과 yscale을 지정해주어 각 축의 스케일을 조정할 수 있다. (linear, log, symlog등~)

```python
# 그리드 설정
x = np.linspace(0, 10, 100, 1000)
y = x ** 3

plt.plot(x,y)
plt.yscale('log')
plt.grid(True,color='lime',linestyle=":",alpha=0.5,)
plt.show()
```

plt.grid() 함수로 그리드를 설정할 수 있다. 

또한 color, alpha, linestyle 을 설정해주어 그리드의 스타일을 변경할 수 있다.

```python
# 타이틀 설정

x = np.linspace(0, 10, 100, 1000)
y = x ** 3

plt.plot(x,y)
plt.title("Title")
plt.show()
```

plt.title() 함수로 제목을 설정할 수 있고 fontsize, fontweight, loc, pad 등으로 폰트를 조정할 수 있다

```python
# 눈금 표시하기

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), ('Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June',''))
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

```python
# 산점도 그리기
np.random.seed()
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y, s=50,c=np.random.rand(n),alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

plt.scatter를 이용해 산점도 그래프를 그릴 수 있다

s는 사이즈, c는 컬러를 말하며 cmap은 색상범위?인거같고 alpha는 투명도 옆에 막대는 colorbar

```python
# 히스토그램 그리기
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.hist((weight,weight2),bins=10, histtype='bar')
plt.show()
```

plt.hist 함수를 이용해 히스토그램을 그릴 수 있고 bins는 구간갯수 를 의미한다

```python
# 에러바 그리기
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [2.3, 3.1, 1.7, 2.5] #오차

plt.errorbar(x, y, yerr=yerr)
plt.show()
```

```python
# 파이차트 그리기
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.pie(ratio, labels=labels, autopct='%.1f%%', wedgeprops=wedgeprops)
plt.show()
```

```python
# 여러개의 그래프 그리기
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

subplot(nrows, ncols, index) 형태로 함수를 여러개 그릴 수 있다

형태는 이미지 참조, sharex=True를 이용해 중복된 축을 공유할 수 있다.

```python
# rcParams 사용하기
plt.rcParams['figure.figsize'] = (6, 3) # 그래프 이미지 크기
plt.rcParams['font.size'] = 12          #폰트 사이즈
plt.rcParams['lines.linewidth'] = 3     #선 두께
plt.rcParams['lines.linestyle'] = '-'   #선 종류

plt.plot([1, 2, 3, 4], [4, 6, 2, 7])
plt.show()
```

rcParams는 종류가 매우 다양하다 이 파라미터로 그래프를 세세하게 조정할 수 있다.

```python
# 박스플롯 그리기
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

np.random.seed()
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5)
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

plt.boxplot 함수를 이용해 박스플롯을 그릴수 있다.

```python
# 바이올린플룻 그리기
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

np.random.seed()
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')


plt.show()
```

```python
# 객체 지향 인터페이스 (커스터마이징)
plt.rcParams['figure.figsize']=(10,5)
x = np.arange(1, 5)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
ax[0][0].plot(x, np.sqrt(x))      # left-top
ax[0][1].plot(x, x)               # right-top
ax[1][0].plot(x, -x+5)            # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5))   # right-bottom
plt.show()
```

위와같이 ax를 지정해주고 추가로 레이블을 넣거나 그래프를 꾸미거나 해주면 된다.

ax.set_~~  함수로 그래프 조정 가능

```python
# 두종류의 그래프 그리기
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price ($)')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx() # X축을 공유하는 이중Y축 만들기
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18)
ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')

plt.show()
```
